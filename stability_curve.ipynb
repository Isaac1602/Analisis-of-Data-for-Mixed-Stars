### On this part of the code it is created a dataframe in pandas

```python
import pandas as pd
```

```python
# Create a list of names for the header

column_names = ["rho0", "phi0", "M_tail", "R_99", "Rstarb", "Rstarf", "NoPartB",
                "NoPartF", "omega", "lambda", "kapa", "gama", "emu"]

# Create a DataFrame with the specified column names and empty data
inestable_mases = pd.DataFrame(columns=column_names)

# Print the DataFrame
print(inestable_mases)
```

### This section defines two functions that search the inestable mases

```python
def search_boson (path, inestable_mases):

    #Read the data
    df = pd.read_csv(path,  delim_whitespace=True, header=None, names=column_names)
    
    #Find the minimum value on the number of Bosons
    min_value_NoPartB = df['NoPartB'].min()
    
    # Extracts the row with minimum value
    configuracion_inestable = df[df['NoPartB'] == min_value_NoPartB]
    
    inestable_mases = pd.concat([inestable_mases, configuracion_inestable], ignore_index=True)
    return inestable_mases
```

```python
def search_fermion (path, inestable_mases):

    #Read the data
    df = pd.read_csv(path,  delim_whitespace=True, header=None, names=column_names)

    #Find the minimum value on the number of Bosons
    min_value_NoPartF = df['NoPartF'].min()
    
    # Extracts the row with minimum value
    configuracion_inestable = df[df['NoPartF'] == min_value_NoPartF]

    inestable_mases = pd.concat([inestable_mases, configuracion_inestable], ignore_index=True)
    return inestable_mases
```

#### The following snippet of code is used to search over all the datasets

```python
i = 0
while i == 0: 
    path = input("Write the name: ")
    if "B" in path:
       inestable_mases = search_boson(path, inestable_mases)
    elif "F" in path:
       inestable_mases =  search_fermion(path, inestable_mases)
    else:
        i = 1
        continue
    i = int(input("Select 0 for more data or 1 to finish analysing data: "))
    if i == 0:
        print("The process must continue\n")
    elif i ==1:
        print("The process must stop \n")

inestable_mases

    
```

### It is needed to import matplotlib o seaborn to graph the data

```python
# Extract the columns you want to plot
x_column = 'rho0'  # Replace with the actual column name for the x-axis
y_column = 'phi0'  # Replace with the actual column name for the y-axis

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(data[x_column], data[y_column], marker = 'o', color = "green")

# Set labels and title
ax.set_xlabel('Rho')
ax.set_ylabel('Phi')
ax.set_title('Scatter Plot of ' + 'Rho' + ' vs. ' + 'Phi')

# Customize the plot (optional)
plt.grid(True)
plt.xlim(min(data[x_column]), max(data[x_column]+0.002))
plt.ylim(min(data[y_column]), max(data[y_column]+0.02))

# Save the plot
plt.savefig("Simple_figure.png")

# Show the plot
plt.show()


```
